```python
#X=features,y=target
#X_train y_train    70%
#X_train  y_train   30%
X=df.drop('species', axis=1)
y=df['species']
print(X,y)
```

<h3>DECISION TREE CLASSIFIER 

```python

#cmd-pip install scipy
#cmd-pip install scikit-learn
#cmd-pip install scikit-learn --upgrade
#              nameofmodule
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
#spliting data
#random_state 100 .shuffle first 100 with remaning data or else it will take first 100 only
#test_size0.3=30%
```

```python
X_train.shape
#70%....4 features
```

```python
y_train.shape
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
print("The shape of X_test ",X_test.shape,"the shape of y_test ",y_test.shape)
#30%
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model= DecisionTreeClassifier()
#tree
```

```python
model.fit(X_train,y_train)
#traning model
```

```python
#predict y_test by reffering X_test
y_pred = model.predict(X_test)
```

```python
y_pred
```

evaluation of accuracy score of the model using matrix

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

95% accurate
